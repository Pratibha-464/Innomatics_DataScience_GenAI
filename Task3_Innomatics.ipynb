<b>Question 1</b>:Given the array of integers nums, you will choose two different indices i and j of that array. Return the maximum value of (nums[i]-1)*(nums[j]-1).

```python
nums = [3,4,5,2]
max_val = 0

for i in range(len(nums)):
    for j in range(len(nums)): 
        if i != j:
            a = (nums[i]-1)*(nums[j]-1)
            if a > max_val:
                max_val = a


print("Greatest value is:", max_val,)
```

<b>Question 2</b>:There are n soldiers standing in a line. Each soldier is assigned a unique rating value.
You have to form a team of 3 soldiers amongst them under the following rules:
Choose 3 soldiers with index (i, j, k) with rating (rating[i], rating[j], rating[k]).
A team is valid if: (rating[i] < rating[j] < rating[k]) or (rating[i] > rating[j] > rating[k]) where (0 <= i < j < k < n).
Return the number of teams you can form given the conditions. (soldiers can be part of multiple teams).

```python
rating = [2,5,3,4,1]
n = len(rating)

count = 0

for i in range(n):
    for j in range(i+1, n):
        for k in range(j+1, n):
            if rating[i] < rating[j] < rating[k] or rating[i] > rating[j] > rating[k]:
                print (rating[i], rating[j], rating[k])
                count += 1
                
print("The number of teams you can form:", count)                
```

<b>Question 3</b>:Given two integer arrays startTime and endTime and given an integer queryTime.

Return the number of students doing their homework at time queryTime. More formally, return the number of students where queryTime lays in the interval [startTime[i], endTime[i]] inclusive.

<b>Question 4</b>:Given an integer num, return the number of steps to reduce it to zero.
In one step, if the current number is even, you have to divide it by 2, otherwise, you have to subtract 1 from it.

```python
num = 123
count = 0

while num > 0:
    if num % 2 == 0:
        num = num // 2
        
    else:
        num -= 1
    count += 1    

print ("The result of the number", count)
```

<b>Question 5</b>:Given an integer n, return an array ans of length n + 1 such that for each i (0 <= i <= n), ans[i] is the number of 1's in the binary representation of i.

```python
n  =  int(input('Enter a lenght if the array:'))
ans = [0]* (n + 1)

for i in range(1, n+1):
    ans[i] = ans[i // 2] + (i % 2)

print (ans)
```
