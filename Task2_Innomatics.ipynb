<b>Question 1</b>: Given a valid (IPv4) IP address, return a defanged version of that IP address.
A defanged IP address replaces every period "." with "[.]".

```python
address = "255.100.50.0"
result = ""

for i in address :
    if i == "." :
        result += "[.]"

    else:
        result += i

print (result)
```

<b>Question 2</b>: Given an array nums of integers, return how many of them contain an even number of digits.

```python
arr = [12,345,2,6,7896]

lengths = []

for i in arr:
    count = 0
    n = i

    if n == 0:     
        count = 1
    else:
        while n > 0:
            count += 1
            n //= 10

    if count % 2 == 0:   
        lengths.append(count)
        print (i, "has a even digits")
```

<b>Question 3</b>: Given an array of integers nums, return the number of good pairs.
A pair (i, j) is called good if nums[i] == nums[j] and i < j.

```python
nums = [1, 2, 3, 1, 1, 3]
count = 0

for i in range(len(nums)):
    for j in range(i + 1, len(nums)):
        if nums[i] == nums[j]:
            count += 1
            print("Good pair:", i, j)

print("Total pairs:", count)
```

<b>Question4: </b>: Given the array nums, for each nums[i] find out how many numbers in the array are smaller than it. That is, for each nums[i] you have to count the number of valid j's such that j != i and nums[j] < nums[i].
Return the answer in an array.

```python
lst = [8,1,2,2,3]
result = []

for i in range(len(lst)):
    count = 0
    for j in range(len(lst)):
            if  lst[j] < lst[i] and j != i:
                count += 1
    result.append(count)

print(result)
```

<b>Question 5</b>: Given an integer number n, return the difference between the product of its digits and the sum of its digits.

```python
def difference_function(num):
    
    total = 0
    prod =  1
    while num > 0:
        l = num % 10   
        total += l
        prod *= l
        num = num // 10      
    
    return  prod - total
```

```python
print ("The final output is", difference_function(765))
```

<b>Question 6</b>:You are given an integer n and an integer start.
Define an array nums where nums[i] = start + 2 * i (0-indexed) and n == nums.length.
Return the bitwise XOR of all elements of nums.

```python
n = int(input('Enter a number for the lenght of the list:'))
start = int(input('Enter a start value:'))
nums = []

for i in range(n):
    a = start + 2 * i
    nums.append(a)
print (nums)

result = 0
for num in nums:
    result ^= num

print ("Bitwise XOR of all elements of nums", result)
```
