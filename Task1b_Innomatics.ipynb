<b>Question 1</b>: Given an array nums. We define a running sum of an array as runningSum[i] = sum(nums[0]â€¦nums[i]).
Return the running sum of nums. 

```python
nums = [int(x) for x in input("Enter numbers for the list: ").split(',')]
snum = []

count = 0
for i in nums:
    count += i
    snum.append(count)
    
print(snum)
    
```

<b>Question 2</b>: Given the array nums consisting of 2n elements in the form [x1,x2,...,xn,y1,y2,...,yn].
Return the array in the form [x1,y1,x2,y2,...,xn,yn].

```python
import random
n = int(input('Enter a number:'))

lst = list(random.sample(range(1, 500), 2*n))
print(lst)
new_lst = []
used = set()

while len(new_lst) < len(lst):
    i = random.randint(0, len(lst)-1)
    if i not in used:
        used.add(i)
        new_lst.append(lst[i])

print(new_lst)
```

<b>Question 3</b>: There are n kids with candies. You are given an integer array candies, where each candies[i] represents the number of candies the ith kid has, and an integer extraCandies, denoting the number of extra candies that you have.

Return a boolean array result of length n, where result[i] is true if, after giving the ith kid all the extraCandies, they will have the greatest number of candies among all the kids, or false otherwise.

Note that multiple kids can have the greatest number of candies.

```python
candies = [int(x) for x in input("Enter numbers candies of each kids: ").split(',')]
extraCandies = int(input('Enter a number of extracandies:'))
totCandies = []
maxCandies = max(candies)

for i in range(len(candies)):
    a = candies[i] + extraCandies
    totCandies.append(a >= maxCandies)

print(totCandies)
```
